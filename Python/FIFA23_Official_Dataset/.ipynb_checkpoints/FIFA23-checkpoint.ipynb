```python
df5.head()
```

```python
# Define a function to convert 'M' notation to numeric values
def convert_million_and_thousand(value):
    multiplier = 1
    if 'M' in value:
        multiplier = 1000000
    elif 'K' in value:
        multiplier = 1000
    
    return float(value[:-1]) * multiplier if value[-1] in ['M', 'K'] else float(value)
```

```python
# Apply the function to the 'Value' column
df5['Value'] = df5['Value'].apply(convert_million_and_thousand)
```

```python
df5.head()
```

```python
df5['Wage'] = df5['Wage'].apply(convert_million_and_thousand)
```
